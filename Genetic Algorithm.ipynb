```python
def InitialPopulation():
    
    Population = [] # This stores the initial population 

    # The loop below initializes the List with N 8-bit chromosomes

    for n in range(N): # loop which will dynamically intialize a list of size N with N number of populations
    
        for i in range(N): # will run for population size , generating N number of chromosomes 
    
            Chromosomes = [] 
        
            for j in range(0, L): # range specified for 8-bit chromosome
                x = str(random.randint(0, 1)) # only 0's and 1's for binary
                Chromosomes.append(x) # appending bit by bit to the Chromosomes List

            Chromosome = ''.join(Chromosomes) # converting the list to a string
        
        Population.append(Chromosome) # chromosome is appended to the Population


    return(Population)
```

```python
                        # The code below is the implementation of the fitness function phase : #
    

def FitnessFunction(Population):

    fitnessvalue = []

    for chromosome in Population :
        fitnessvalue.append(int(chromosome,2)) # converting the binary string into an integer
        if L == 8:
            if 255 in fitnessvalue:    # for an 8 bit binary number highest value is 255
                print("Ideal case reached.")# we stop once we reach the ideal case
                return 0 
        elif L == 9:
            if 511 in fitnessvalue:    # for a 9 bit binary number highest value is 511
                print("Ideal case reached.")# we stop once we reach the ideal case
                return 0
        else:
            if 1023 in fitnessvalue:    # for a 10 bit binary number highest value is 1023
                print("Ideal case reached.")# we stop once we reach the ideal case
                return 0 
                                  
        
    # using Selection Sort to sort the population in ascending order of fitness function 
    
    print("Fitness Values : ",fitnessvalue)
    
    for i in range(len(fitnessvalue)): 
       
        Min = i 
        for j in range(i+1, len(fitnessvalue)): 
            if fitnessvalue[Min] > fitnessvalue[j]: 
                Min = j 
            
        fitnessvalue[i], fitnessvalue[Min] = fitnessvalue[Min], fitnessvalue[i] 
        Population[i], Population[Min] = Population[Min], Population[i]
    
    return(Population)
```

```python

                            # The code below is the implementation of the selection phase : #
    
def Selection(Population):
    
    SelectedChromosomes = []

    #  we will be Selecting k-1 best states from the population
    for i in range (N-1):
        SelectedChromosomes.append(Population[i])
    
    return(SelectedChromosomes)
```

```python
                        
    
                        # The code below is the implementation of the cross over phase : #
        
def CrossOver(SelectedChromosomes):
    
    CrossedOverPopulation = [] 

    for i in range(len(SelectedChromosomes)):
        CrossedOverPopulation.append(SelectedChromosomes[i])
    
    #As in this phase from the selected chromosomes we have to again produce N number of chromosomes 
    # so for combination making we will duplicate the first one of the selected chromosomes
    
    CrossedOverPopulation.append(SelectedChromosomes[0]) 

#print(CrossedOverPopulation)

    # storing the copy of the last selected chromosome as we will use it twice for combination making

    copy = CrossedOverPopulation[(len(CrossedOverPopulation))-1] 

    x=0 # where x is index of an L-bit chromosome

    while x < len(CrossedOverPopulation):  
    
        if x+1 < len(CrossedOverPopulation): # check for the last chromosome in the list
       
        # We will take consider 5th bit as the crossover point
      # first we will split the chromosome into two parts (bit 1 to 5 and bit 6 to last bit)
      # and then concatenate it with part of another chromosome to form a new chromosome
    
            c1 = (CrossedOverPopulation[x])[:5]
            c2 = (CrossedOverPopulation[x])[5:L]
            c3 = (CrossedOverPopulation[x+1])[:5]
            c4 = (CrossedOverPopulation[x+1])[5:L]
            CrossedOverPopulation[x] = c1+c4
            CrossedOverPopulation[x+1] = c3+c2
    
        else:  # this will work for odd number of population 
        
            c1 = (CrossedOverPopulation[x])[:5]
            c2 = (CrossedOverPopulation[x])[5:L]
            c3 = (copy)[:5]
            c4 = (copy)[5:L]
            CrossedOverPopulation[x] = c3+c2
    
        x += 2
    
    return(CrossedOverPopulation) 
```

```python
                    
    
                        # The code below is the implementation of the mutation phase : #

def Mutation(CrossedOverPopulation):
    
    # For each chromosome we will select a random bit and change its value from 0 to 1 
    #print("M")
    NewPopulation = []

    for chromosome in CrossedOverPopulation :
        while True:  
            i = random.randrange(L-1)   # a random bit is choosen for mutation
            if (chromosome[i] == '0'): # keep finding until the value of the bit is not 0
                break
        chromosome = chromosome[:i] + '1' + chromosome[i + 1:]
        NewPopulation.append(chromosome)
    
    return(NewPopulation)
```

```python
import random 

N = int(input("Enter the population size: ")) # Here "N" is taken to be the size of population
L = int(input("Enter the length of chromosome (8 bit - 10 bit): ")) # Here "L" is taken to be the chromosome length
G = 0   # This is the number of generations 

initial = InitialPopulation()       
print("\nInitial Population : ",initial)
population = FitnessFunction(initial)
selectedchromosomes = Selection(population)
crossedoverpopulation = CrossOver(selectedchromosomes)
newpopulation = Mutation(crossedoverpopulation)
print("\nPopulation : ",newpopulation)

while True :
    
    Population = FitnessFunction(newpopulation)
    if Population == 0:   # if the ideal case is reached the Fitness Function returns zero
        break             # we stop once we get the ideal case 
    SelectedChromosomes = Selection(Population)
    CrossedOverpopulation = CrossOver(SelectedChromosomes)
    newpopulation = Mutation(CrossedOverpopulation)
    print("\nPopulation : ",newpopulation)
    G += 1
    
print("\nNo of Generations : ",G)
```
